### Molecular Hydrogen Generally and it's Importance in the Universe

Molecular hydrogen (H₂) is the simplest molecule, composed of two hydrogen atoms, and was the first neutral molecule formed in the universe. Additionally, it is the most common molecule in the universe and plays a key role in the formation of stars and galaxies.

In the ISM (Interstellar Medium), molecular hydrogen is the primary component of cosmic molecular clouds. These clouds are large, dense regions and are among the coldest objects in the universe, with temperatures ranging from 10 to 50 Kelvin. In the coldest and densest corners of these clouds, star formation occurs. H₂ is a highly effective coolant at the low temperatures found in molecular clouds, which is essential as it allows the gas to compress further, eventually reaching densities that enable gravitational collapse and lead to star formation.

### Formation of Molecular Hydrogen
Although H₂ is the most abundant molecule in space, its formation is complex and not easily maintained. The issue with molecular hydrogen is that it is vulnerable to high-energy ultraviolet (UV) radiation, especially at wavelengths around 912-1100 angstroms (~90-110 nm). When H₂ molecules absorb UV photons, they are excited to higher energy states. If enough energy is absorbed, the bond between the two hydrogen atoms breaks, releasing them back into atomic form in a process known as photodissociation. This vulnerability creates the need for some form of protection against UV radiation to maintain stable levels of H₂. Additionally, the formation of molecular hydrogen itself is inefficient in the ISM without dust particles acting as catalysts.

##### Shielding of Molecular Hydrogen
A natural shield is provided by higher densities found in cosmic clouds, where a density greater than 100 particles/cm³ is necessary to counterbalance the rate of photodissociation. Dust grains—tiny solid particles composed of elements such as carbon, silicon, and metals—are concentrated in dense regions like cosmic clouds and act as an effective shield for molecular hydrogen. These particles are highly efficient at absorbing and scattering UV radiation, thus surrounding molecular hydrogen and absorbing much of the harmful UV light before it reaches the H₂ molecules.
But, H₂ can protect itself as well through a process called self-shielding, in which the outer layers of a molecular cloud absorb UV radiation and undergo dissociation, preserving the inner layers of H₂ from exposure.

##### H₂ Formation on Dust Particles
When a hydrogen atom collides with a dust grain and adheres to its surface, it can migrate across the surface to find another hydrogen atom. When two hydrogen atoms meet on the grain surface, they combine to form molecular hydrogen (H₂). This bonding process releases a small amount of energy, which is absorbed by the dust grain and dissipated as heat. The newly formed H₂ molecule then gains enough energy to escape from the dust particle.

##### Gas-Phase Reactions 
In very low-density regions, where dust grain catalysis is inefficient, alternative gas-phase reactions can form H₂, though this process is far less efficient and much rarer in the ISM. In this process, a proton (hydrogen ion)—usually created by cosmic rays—combines with a hydrogen atom to form an H₂⁺ ion. This H₂⁺ ion then reacts with either a free electron in the ISM or another hydrogen atom, resulting in the formation of neutral molecular hydrogen and a hydrogen ion.

Note: H₂ cannot generally form in the ISM through the direct collision of two hydrogen atoms. The reason behind this is that when two hydrogen atoms combine, energy is released as they transition from an excited state to a lower bound state. For the H₂ molecule to remain stable, this energy must be released or transferred; otherwise, the molecule will break apart again. In the ISM, there is virtually no means of dissipating this energy. When H₂ forms on a dust particle, the energy can be discharged into the grain, which makes dust essential for the formation of molecular hydrogen

### Distribution of Molecular Hydrogen in the Milky Way
For understanding galactic evolution, molecular hydrogen is crucial. The amount and distribution of H₂ in a galaxy determine the galaxy's capacity to form new stars. Galaxies rich in H₂ are often actively forming stars, while galaxies with little H₂ tend to be older, less active, and often “quenched” in terms of star formation. Since star formation occurs primarily in cosmic clouds, these clouds are naturally the main locations where molecular hydrogen can be found. Generally, cosmic clouds are located in regions with increased density and mass. In a spiral galaxy like the Milky Way, this typically occurs in the spiral arms. The highest concentrations of H₂ are found in the inner regions of the galaxy, within about 5 kpc of the galactic center, where the disk is densest.

### Why is it so Hard to Dedect H₂?
Due to many reasons, we would really like to know the exact distribution of molecular hydrogen clouds in our galaxy and others as well. Interestingly, in most cases, it is impossible for us to detect H₂ directly. The main issues are:

1. H₂ is a symmetrical molecule and lacks a permanent dipole moment, which prevents it from emitting or absorbing light in the radio and infrared wavelengths commonly used to observe other molecules.

2. Although H₂ can technically emit through quadrupole transitions (a higher-order effect), these transitions are extremely weak and occur only under specific conditions, such as high temperatures (over 100 K), where the molecule has enough energy to enter an excited state. But, these temperatures are not typically found in classical cosmic clouds.

3. H₂ can absorb UV light, and these absorptions can, in theory, be used to detect it. However, UV light is mostly absorbed by interstellar dust and gas, as explained earlier, making it difficult to observe H₂ directly in dense molecular clouds.
