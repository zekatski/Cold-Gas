### Molecular Hydrogen Generally and its Importance in the Universe

Molecular hydrogen ($H_2$) is the simplest molecule, composed of two hydrogen atoms, and was the first neutral molecule formed in the universe. Additionally, it is the most common molecule in the universe and plays a key role in the formation of stars and galaxies.

In the ISM (Interstellar Medium), molecular hydrogen is the primary component of cosmic molecular clouds. These clouds are large, dense regions and are among the coldest objects in the universe, with temperatures ranging from 10 to 50 Kelvin. In the coldest and densest corners of these clouds, star formation occurs. $H_2$ is a highly effective coolant at the low temperatures found in molecular clouds, which is essential as it allows the gas to compress further, eventually reaching densities that enable gravitational collapse and lead to star formation.

### Formation of Molecular Hydrogen
Although $H_2$ is the most abundant molecule in space, its formation is complex and not easily maintained. The issue with molecular hydrogen is that it is vulnerable to high-energy ultraviolet (UV) radiation, especially at wavelengths around 912-1100 angstroms (~90-110 nm). When $H_2$ molecules absorb UV photons, they are excited to higher energy states. If enough energy is absorbed, the bond between the two hydrogen atoms breaks, releasing them back into atomic form in a process known as photodissociation. This vulnerability creates the need for some form of protection against UV radiation to maintain stable levels of $H_2$. Additionally, the formation of molecular hydrogen itself is inefficient in the ISM without dust particles acting as catalysts.

##### Shielding of Molecular Hydrogen
A natural shield is provided by higher densities found in cosmic clouds, where a density greater than 100 particles/cm¬≥ is necessary to counterbalance the rate of photodissociation. Dust grains‚Äîtiny solid particles composed of elements such as carbon, silicon, and metals‚Äîare concentrated in dense regions like cosmic clouds and act as an effective shield for molecular hydrogen. These particles are highly efficient at absorbing and scattering UV radiation, thus surrounding molecular hydrogen and absorbing much of the harmful UV light before it reaches the $H_2$ molecules.
But, $H_2$ can protect itself as well through a process called self-shielding, in which the outer layers of a molecular cloud absorb UV radiation and undergo dissociation, preserving the inner layers of $H_2$ from exposure.

##### $H_2$ Formation on Dust Particles
When a hydrogen atom collides with a dust grain and adheres to its surface, it can migrate across the surface to find another hydrogen atom. When two hydrogen atoms meet on the grain surface, they combine to form molecular hydrogen ($H_2$). This bonding process releases a small amount of energy, which is absorbed by the dust grain and dissipated as heat. The newly formed $H_2$ molecule then gains enough energy to escape from the dust particle.

##### Gas-Phase Reactions 
In very low-density regions, where dust grain catalysis is inefficient, alternative gas-phase reactions can form $H_2$, this process though is far less efficient and much rarer in the ISM. In this process, a proton (hydrogen ion) ‚Äî usually created by cosmic rays ‚Äî combines with a hydrogen atom to form an $H_2‚Å∫$ ion. This $H_2‚Å∫$ ion then reacts with either a free electron in the ISM or another hydrogen atom, resulting in the formation of neutral molecular hydrogen and a hydrogen ion.

Note: $H_2$ cannot generally form in the ISM through the direct collision of two hydrogen atoms. The reason behind this is that when two hydrogen atoms combine, energy is released as they transition from an excited state to a lower bound state. For the $H_2$ molecule to remain stable, this energy must be released or transferred; otherwise, the molecule will break apart again. In the ISM, there is virtually no means of dissipating this energy. When H‚ÇÇ forms on a dust particle, the energy can be discharged into the grain, which makes dust essential for the formation of molecular hydrogen.

### Distribution of Molecular Hydrogen in the Milky Way
For understanding galactic evolution, molecular hydrogen is crucial. The amount and distribution of $H_2$ in a galaxy determine the galaxy's capacity to form new stars. Galaxies rich in $H_2$ are often actively forming stars, while galaxies with little $H_2$ tend to be older, less active, and often ‚Äúquenched‚Äù in terms of star formation. Since the formation of $H_2$ occurs primarily in cosmic clouds, these clouds are naturally the main locations where molecular hydrogen can be found. Generally, molecular clouds are located in regions with increased density and mass. In a spiral galaxy like the Milky Way, this typically occurs in the spiral arms. The highest concentrations of $H_2$ are found in the inner regions of the galaxy, within about 5 kpc of the galactic center, where the disk is densest.

In figure 2, we can clearly identify the galactic plane as a horizontal line, the spiral arms on either side (left and right corners), and the significantly increased density in the central region of the galaxy.

### Why is it so Hard to Detect $H_2$?
Due to many reasons, we would really like to know the exact distribution of molecular hydrogen clouds in our galaxy and others as well. Interestingly, in most cases, it is impossible for us to detect $H_2$ directly. The main issues are:

1. $H_2$ is a symmetrical molecule and lacks a permanent dipole moment, which prevents it from emitting or absorbing light in the radio and infrared wavelengths commonly used to observe other molecules.

2. Although $H_2$ can technically emit through quadrupole transitions (a higher-order effect), these transitions are extremely weak and occur only under specific conditions, such as high temperatures (over 100 K), where the molecule has enough energy to enter an excited state. But, these temperatures are not typically found in classical cosmic clouds.

3. $H_2$ can absorb UV light, and these absorptions can, in theory, be used to detect it. However, UV light is mostly absorbed by interstellar dust and gas, as explained earlier, making it difficult to observe $H_2$ directly in dense molecular clouds.

### The History of the Hunt for $H_2$ ###

#### Early Works and Ideas ####

In 1904, Johannes Hartmann observed the spectral lines of ionized calcium (Ca II) that appeared to remain unchanged despite the star‚Äôs motion. This finding led him to conclude that there was something between the light source and Earth that absorbed photons at a specific wavelength. The paper he published marks the first scientific evidence of the existence of clouds in the ISM. Later works of the early 20th century suggested that these interstellar clouds also consist of molecular components. In the following decades, further publications provided viable evidence for this idea. During the 1950s, the 21 cm line in radio spectra caused by atomic hydrogen was discovered, leading astronomers to search for a similar tracer of molecular hydrogen. The abundance of $H_2$ clouds was theorized in a paper published in 1951, but this very paper also discussed the potential difficulty of finding molecular hydrogen due to its physical properties.

#### Searching for Effective Tracers ####

Almost two decades later, in the late ‚Äô60s, new publications aimed to encourage scientists to shift their research objectives in a different direction. The development of an indirect $H_2$ tracing method was necessary. In the following years, a concept for the process of $H_2$ formation on dust particles was proposed, further supporting the idea that, in interstellar space, hydrogen is also abundant in its molecular form. During the ‚Äô70s, many new ideas and discoveries emerged, starting with a 1970 paper on observations of line radiation attributed to CO in cold interstellar gas, which introduced the molecule as a viable $H_2$ tracing candidate. This idea was successfully reinforced by papers published in the following years as CO was established as the primary tracer for molecular hydrogen. A notable theoretical advancement from this time is the CO-to-$H_2$ conversion factor, commonly denoted as XCO. The classic review paper on the evolution of CO as a tracer was published in 1986, providing an interesting view of the historical development of carbon monoxide as a tracer. Over the years, these carbon-monoxide-centered methods contributed to the search for interstellar molecular clouds and were continuously improved and refined until the 2000s. A more recent paper from 2013 gives a comprehensive review of CO as a tracer of $H_2$, also discussing the limitations and dependencies of this method. These limitations, mostly occurring in CO ‚Äúdark‚Äù regions, have prompted a search for CO-unrelated tracing techniques.

#### Modern Methods and Alternatives to CO ####

With newer and more advanced instrumentation, the accuracy and range of modern surveys have drastically increased each year. This has resulted in a wider spectrum of available $H_2$ tracing methods, as well as an improved understanding of the role CO holds in locating molecular hydrogen. This more recent data has been gathered using a myriad of instruments and techniques. While radio and (sub)millimeter surveys manage to map molecular tracers other than CO, which are particularly helpful in regions of low CO abundance, infrared observations can locate dust emissions, which are also suggested indicators of $H_2$ presence. Additionally, studies in the gamma-ray range allow for the detection of dark gas clouds interacting with cosmic rays. A paper published in the 2015 Annual Review of Astronomy and Astrophysics provides a broad discussion and analysis of the Milky Way‚Äôs molecular clouds and addresses the various methods of tracing $H_2$ and the effectiveness of these tracers.


### Summary Tracer Methods


Using CO as a Proxy: 
Carbon monoxide (CO) is often used as a tracer for molecular hydrogen because it emits strongly in the radio wavelength. Particularly at the J=1‚Üí0J=1‚Üí0 rotational transition of 2.6 mm and thus is easy to detect. 
Since CO and $H_2$‚Äã often coexist in dense molecular clouds,astronomers use the ratio between CO intensity and $H_2$‚Äã column density, known as the CO-to-$H_2$‚Äã conversion factor (or XCOXCO‚Äã), to estimate $H_2$ densities. 
The value of XCOXCO‚Äã can vary with cloud conditions, such as metallicity,density and radiation environment.



UV Absorption Lines:
In regions near hot stars, molecular hydrogen can be detected through ultraviolet absorption lines in the Lyman and Werner bands, which occur around 912‚Äì1100 √Ö. 
Observations from instruments like the Hubble Space Telescope and the Far Ultraviolet Spectroscopic Explorer (FUSE) have been crucial in directly detecting $H_2$‚Äã via this method, especially in translucent and diffuse regions of the ISM. They found that the $H_2$ column density varies widely, reflecting differneces in olecular cloud density, UV radiation levels, and gas temperature across the ISM.

Infrared Emission Lines: 
Vibrational and rotational transitions of $H_2$ in the near-infrared, particularly at wavelengths of 2.12 ¬µm, can be used to study warmer molecular hydrogen, such as in photodissociation regions (PDRs) near hot stars or shocked gas. $H_2$ infrared emission can provide insight into the conditions of the gas such as temperature distribution.
Instruments like the Spitzer Space Telescope and ground-based observatories equipped with IR spectrometers have provided valuable data on $H_2$‚Äã in these environments

Gamma-Ray Observations: 
Cosmic rays interacting with $H_2$‚Äã in molecular clouds produce gamma-ray emissions that can be detected by telescopes like the Fermi Gamma-ray Space Telescope. 
This method, while indirect, allows estimation of molecular hydrogen in regions where CO is under-abundant, such as in low-metallicity environments where traditional methods may be less effective

$2 \times 10^{20} \,\text{cm}^{-2} (\text{K km s}^{-1})^{-1}$  

Here's a breakdown of what each part means:
 $2 \times 10^{20} \,\text{cm}^{-2}$: This part represents a column density, which is a measure of the number of particles (e.g., hydrogen molecules) per square centimeter along a line of sight. The unit $\text{cm}^{-2}$ indicates the density per square centimeter.
    
 $(\text{K km s}^{-1})^{-1}$: This unit corresponds to the "brightness temperature" scale, often observed in radio astronomy to measure line emissions. The combination of Kelvin (K), kilometers per second (km/s), and the reciprocal exponent means that this expression converts from a measurement of brightness temperature (in K km/s) to column density (in particles per cm $^{-2}$).

In simpler terms, this factor  $2 \times 10^{20} \,\text{cm}^{-2} (\text{K km s}^{-1})^{-1}$  is often used to estimate the amount of molecular hydrogen present along a given line of sight based on observed emission from carbon monoxide (CO), which is an indirect tracer of hydrogen molecules.


Integration over velocity: By integrating this intensity over the velocity (in km/s), we obtain the area under the line, which gives ùëä(N2H+). This area represents the total radiative power of the molecules along the line of sight across all velocities, but without a direct conversion into an energy unit.

In summary:
ùëä(N2H+)
W(N2H+) indicates the total "amount" of radiation emitted by N2H+ molecules along the line of sight, integrated over the velocities, in the unit K/km*s^-1.
Thus, it is more a measure of the total emission of these molecules in the cloud, but not the energy per unit time.

## Confidence Intervals

For this exercise, we chose to plot the X-factor for both CO isotopologues $\text{C}^{18}\text{O}$ and $^{13}\text{CO}$ as it is a combined quantity already provided in the raw data. Interestingly, the distributions of the two conversion factors differ fundamentally in shape, which led us to fit different types of models to their respective histograms. To improve the analysis, we applied a constraint to both datasets, including only values greater than zero and excluding extreme outliers. This adjustment, however, affected the compatibility with the fitted plots ‚Äî particularly for the $X(\text{C}^{18}\text{O})$-factor. To address this, we included an additional figure showing the $X(\text{C}^{18}\text{O})$ data with negative values retained for comparison.

After computing the first two moments for the two datasets, we determined the 2.5-sigma confidence interval for the mean, which is given by:

$
\left( \bar{X} - Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}},  \bar{X} + Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}} \right)
$

This interval covers nearly 99% of the possible values for the true mean.

```python
from scipy.stats import norm

# plot
plt.figure(figsize=(12, 7))

# data filtering
pv_XC18O = (df['X(C18O)'] >= 0) & (df['X(C18O)'] <= 0.8e-6)  
data_C18O = df['X(C18O)'][pv_XC18O] 

# Histogramm
plt.hist(data_C18O, bins=100, density=True, color='skyblue', edgecolor='skyblue', alpha=0.8, label='Histogram')

# Gaussian distribution
mean_C18O = np.mean(data_C18O)  # mean
std_C18O = np.std(data_C18O)    # standard deviation
data_C18O_tn = data_C18O.count()
std_error_C18O = std_C18O/(data_C18O_tn**(0.5))
x_C18O = np.linspace(min(data_C18O), max(data_C18O), 1000)  # x-values for PDF
pdf_C18O = norm.pdf(x_C18O, mean_C18O, std_C18O)  # PDF of the normal distribution

# Confidence interval
count_C18O = df['X(C18O)'].count()
lost_data_frc = data_C18O_tn/count_C18O
CI_C18O_l = mean_C18O - 2.5*std_error_C18O
CI_C18O_u = mean_C18O + 2.5*std_error_C18O
plt.axvline(CI_C18O_l, color='purple', linestyle='-', lw=2, label=f'lower bound of CI for the mean = {CI_C18O_l:.2e}')
plt.axvline(CI_C18O_u, color='purple', linestyle='-', lw=2, label=f'upper bound of CI for the mean = {CI_C18O_u:.2e}')

# Plot of guassian distibution
plt.plot(x_C18O, pdf_C18O, color='red', lw=2, label='Gaussian Fit')

# mean and standard deviation lines
plt.axvline(mean_C18O, color='blue', linestyle='--', lw=2, label=f'Mean = {mean_C18O:.2e}')
plt.axvline(mean_C18O + std_C18O, color='darkgreen', linestyle='--', lw=2, label=f'+1œÉ = {mean_C18O + std_C18O:.2e}')
plt.axvline(mean_C18O - std_C18O, color='darkgreen', linestyle='--', lw=2, label=f'-1œÉ = {mean_C18O - std_C18O:.2e}')
plt.axvline(mean_C18O + 2.5*std_C18O, color='lime', linestyle='--', lw=2, label=f'+2.5œÉ = {mean_C18O+ 2.5*std_C18O:.2e}')

# axes and plot titel
plt.title('Distribution of X(C18O) with Gaussian Fit', fontsize=16)
plt.xlabel('X(C18O)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xlim(min(data_C18O), 0.8e-6)

# legend
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(fontsize=12)

# Layout 
plt.tight_layout()
plt.show()
```

**Figure 9:** Our Gaussian fit for this dataset ($X(\text{C}^{18}\text{O})$-factor) appears slightly shifted due to the fact that we altered the range of the datapoints to be distributed within a physically realistic domain, i.e. stricly greater than zero, as densities do not take on negative values.

```python
count_C18O = df['X(C18O)'].count()
lost_data_frc = data_C18O_tn/count_C18O
print(f"By cutting negative values and outliers we get rid of a fraction of {100*lost_data_frc:.2f}% of the original dataset.")
C18O_CI_l = mean_C18O - 2.5*std_error_C18O
C18O_CI_u = mean_C18O + 2.5*std_error_C18O
print(f"The confidence interval for the mean, for a 2.5œÉ range, lies between {C18O_CI_l:.2e} and {C18O_CI_u:.2e}.")
```

## Hypothesis Testing

When examining the molecular cloud, we may assume that $H_2$ is almost evenly distributed among the individual cloud fragments. To investigate this, we focus on two fragments that appear to differ the most: cloud fragments B7 and B211 (see the figure below, $H_2$ Density Distribution of Cloud Fragments, for reference). Cloud fragment B7, in particular, stands out due to its seemingly higher overall $H_2$ column density. This leads us to our hypothesis, which we aim to disprove at a 2-sigma significance level.

$H_0$ : The mean column density of $H_2$ of the cloud fragments b211 and b7 are the same. 

$H_a$ : The mean column density of $H_2$ of the cloud fragments b7 is higher.

Significance level: $2 \sigma$


```python
plt.hist(b211['N(H2)'], bins=50, alpha=0.35, label='Distribution of B211', color='blue')
plt.hist(b7['N(H2)'], bins=50, alpha=0.35, label='Distribution of B7', color='brown')
plt.hist(b213['N(H2)'], bins=50, alpha=0.35, label='Distribution of B213', color='green')
plt.hist(b216['N(H2)'], bins=50, alpha=0.35, label='Distribution of B216', color='black')

plt.xlabel('H2 column density')
plt.ylabel('Frequency')
plt.title('H2 Column Density Distribution of Cloud fragments')
plt.legend()

plt.show()
```

For a better understanding of which parts of the cloud we are looking at, there is a visualisation of the cloud fragments down below.

**Figure 14:** Column density of $H_2$ in cloud fraqment B211

Here we calculate and define all the values we need for the test we want to execute.

```python
b7_mean = np.mean(b7['N(H2)'])
b211_mean = np.mean(b211['N(H2)'])
b7_std = np.std(b7['N(H2)'])
b211_std = np.std(b211['N(H2)'])
n1 = b7['N(H2)'].count()
n2 = b211['N(H2)'].count()

print(f"Mean of cloud fraqmant b7: {b7_mean:.2e}")
print(f"Mean of cloud fraqmant b211: {b211_mean:.2e}")
print(f"Standard deviation of cloud fraqment b7: {b7_std:.2e}")
print(f"Standard deviation of cloud fraqment b211: {b211_std:.2e}")
print(f"Sample size of cloud fraqment b7: {n1}")
print(f"Sample size of cloud fraqment b211: {n2}")
```

```python
plt.hist(b211['N(H2)'], bins=50, alpha=0.35, label='Distribution of B211', color='blue')
plt.hist(b7['N(H2)'], bins=50, alpha=0.35, label='Distribution of B7', color='red')


plt.xlabel('H2 column density')
plt.ylabel('Frequency')
plt.title('H2 Column Density Distribution of Cloud fragments B211 and B7')

plt.axvline(b7_mean, color='red', linestyle='-', lw=2, label=f'Mean of b7 = {b7_mean:.2e}')
plt.axvline(b211_mean, color='blue', linestyle='-', lw=2, label=f'Mean of b211= {b211_mean:.2e}')

plt.legend()
plt.show()
```

**Figure 15:** Histogram showing of the $H_2$ column density distribution of cloud fragments B211 and B7 and their means.

This histogram of the $H_2$ column density distribution for the two cloud fragments, along with their means, suggests a significant difference between them. To formally test this, we use a two-sample t-test with unknown population standard deviations but large sample sizes:

$$T_0 = \frac{\bar{X}-\bar{Y}}{\sqrt{S_1^{2}/n_1 + S_2^{2}/n_2}}$$

```python
T_0 = (b7_mean-b211_mean)/((((b7_std**2)/n1) + ((b211_std**2)/n2))**(1/2))
print(f"T-value: {T_0:.2f}")
```

Z-value for  $2\sigma \approx 2.03$

Because $T_0=25.96>2.03$, $H_0$ can be rejected on a $2\sigma$-significance level and even lot higher then 99.9%.

```python
p_value =   (1 - norm.cdf(abs(T_0)))

print(f"P-value for the executed test statistic: {p_value:.5g}")
```

As indicated by the calculated p-value, there is essentially a 0% probability of observing our current test statistic under the assumption that the null hypothesis is true. Consequently, we reject $H_0$.

The $H_2$ column density can be estimated using $C^{18}O$ as a tracer, with a certain conversion factor. This implies an approximately linear relationship, eliminating the need for data transformation. Consequently, we expect higher H‚ÇÇ densities in regions with higher $C^{18}O$ densities, and similarly, lower values where $C^{18}O$ densities are lower.

```python
filtered_df = df[(df['N(H2)'] >= 0) & (df['N(C18O)'] >= 0)].dropna(subset=['N(H2)', 'N(C18O)']) # negative values filtered out

pv_AV = filtered_df['N(H2)']
pv_C18O = filtered_df['N(C18O)']

plt.figure(figsize=(8, 6))
plt.scatter(pv_C18O, pv_AV, color='b', alpha=0.1, label='N(C18O) vs. N(H2)')
plt.xlabel("N(C18O)") 
plt.ylabel("N(H2)")
plt.xlim(left=0)      
plt.ylim(bottom=0)   
plt.title("Relation between N(H2) and N(C18O) in Cloud L1495")
plt.legend()
plt.grid(False)
plt.show()
```

**Figure 16:** Relation between N(H2) and N(C18O) in cloud L1495.

From the plot above, we can infer that the data exhibits a heteroscedastic tendency, meaning the variance is not uniform. Which means, there is greater variance at higher $C^{18}O$ column densities, with measured values becoming less centered as density increases.

```python
array_H2 = np.array([filtered_df['N(H2)']]) 
array_CO = np.array([filtered_df['N(C18O)']])  
mean_H2_CO = np.mean(array_H2 *array_CO)
mean_H2 = np.mean(array_H2)
mean_CO = np.mean(array_CO)
mean_CO_2 = np.mean(array_CO**2)
n_H2_CO = filtered_df['N(H2)'].count()

B_1 = (mean_H2_CO - (mean_H2*mean_CO))/(mean_CO_2 - mean_CO**2)
B_0 = mean_H2 - B_1*mean_CO

X_values = np.linspace(0, 4.4*10**15, num=10**3) 
Y_values = filtered_df['Y_values'] = B_0 + B_1*filtered_df['N(C18O)']

lin_reg_line = B_0 + B_1*X_values
```

```python
plt.figure(figsize=(8, 6))
filtered_df = df[(df['N(H2)'] >= 0) & (df['N(C18O)'] >= 0)].dropna(subset=['N(H2)', 'N(C18O)'])

pv_H2 = filtered_df['N(H2)']
pv_C18O = filtered_df['N(C18O)']

plt.scatter(pv_C18O, pv_H2, color='b', alpha=0.1, label='N(C18O) vs. N(H2)')
plt.plot(X_values, lin_reg_line, color='r', label='linear relation')
plt.xlabel("N(C18O)") 
plt.ylabel("N(H2)")
plt.xlim(left=0, right=4.4*10**15)      
plt.ylim(bottom=0)   
plt.title("Relation between N(H2) and N(C18O) in Cloud L1495")
plt.legend()
plt.grid(False)
plt.show()
```

**Figure 17:** Linear regression fit for the relationship between N(H2) and N(C18O) in cloud L1495.

However, this suggests that using a heteroscedastic approach for a better linear regression fit may not be necessary.

Additionally, the Pearson correlation coefficient indicates that the calculated fit is reasonably good, as shown below.

```python
# Cov(X,Y)
Cov_CO_H2 = (1/(n_H2_CO - 1))*(((filtered_df['N(C18O)'] - mean_CO)*(filtered_df['N(H2)'] - mean_H2)).sum())
# Corelation coefficient
var_H2 = (1/(n_H2_CO - 1))*(((filtered_df['N(H2)'] - mean_H2)**2).sum())
var_CO = (1/(n_H2_CO - 1))*(((filtered_df['N(C18O)'] - mean_CO)**2).sum())          
r_CO_H2 = Cov_CO_H2/((var_CO**(1/2))*(var_H2**(1/2)))
print(f"The covariance of N(H2) and N(C18O) is: {Cov_CO_H2:.2e}")
print(f"Corelation coefficient: {r_CO_H2:.2f}")
```

Bootstrapping allows us to assess how likely the calculated slope of the linear regression is to accurately estimate the true slope of the underlying linear relationship.

```python
from sklearn.linear_model import LinearRegression

# Parameter
N = n_H2_CO
K = 1000   
alpha = 0.032  

# Bootstrap-Process
bootstrap_estimates = []
for _ in range(K):
    # random sampling with replacement
    indices = np.random.choice(N, size=N, replace=True)
    X_sample = filtered_df['N(C18O)'].iloc[indices].values.reshape(-1, 1)
    y_sample = filtered_df['N(H2)'].iloc[indices].values

    # model adaptation
    model = LinearRegression()
    model.fit(X_sample, y_sample)
    bootstrap_estimates.append(model.coef_[0])  # saving of estimated slope value

# CI calculation
bootstrap_estimates = np.sort(bootstrap_estimates)
lower_bound = np.percentile(bootstrap_estimates, 100 * (alpha / 2))
upper_bound = np.percentile(bootstrap_estimates, 100 * (1 - alpha / 2))

print(f"The one sigma confidence interval for the value of the slope: [{lower_bound:.2e}, {upper_bound:.2e}]")
```

The slope of the linear relationship for the original dataset is 3.66e+06. Since this value lies perfectly in the middle of the confidence interval, it appears to be a reliable estimator, suggesting that the original dataset is a good representative sample‚Äîlikely due to its relatively large size.
