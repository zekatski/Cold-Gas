```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from astropy.io import ascii, fits

T_NH2 = pd.read_csv("av_c18o_n2hp_75.dat", sep=r"\s+")
T_NH2.head()
```

```python
plt.figure(figsize=(9, 6))
sc = plt.scatter(T_NH2["ra"], T_NH2["dec"], c=T_NH2["AV"], cmap='plasma', s=50, alpha=0.7)

plt.colorbar(sc, label="Extiction Value")
plt.xlabel("RA")
plt.ylabel("Dec")
plt.title("Extinction")
plt.grid(False)
plt.show()
```
