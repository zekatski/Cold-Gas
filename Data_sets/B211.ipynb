```python
plt.figure(figsize=(9, 6))
sc = plt.scatter(b211["ra"], b211["dec"], c=b211["AV"], cmap='plasma', s=50, alpha=0.7)

plt.colorbar(sc, label="Extiction Value")
plt.xlabel("RA")
plt.ylabel("Dec")
plt.title("Extinction (C18O)")
plt.grid(False)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(b211['N(C18O)'], bins=45, color='skyblue', edgecolor='black', alpha=0.7)

# Step 3: Customize the plot
plt.title('Histogram of C18O Density Distribution')
plt.xlabel('N(C18O)')
density=False
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


# Histogramme mit Transparenz
plt.hist(b211['AV'], bins=50, alpha=0.35, label='Av of B211', color='blue')
plt.hist(b7['AV'], bins=50, alpha=0.35, label='Av of B7', color='brown')
plt.hist(b213['AV'], bins=50, alpha=0.35, label='Av of B213', color='green')
plt.hist(b216['AV'], bins=50, alpha=0.35, label='Av of B216', color='black')

# Alternativ: Histogramme als Umrisse
# plt.hist(data1, bins=30, histtype='step', label='Datensatz 1', color='blue', linewidth=1.5)
# plt.hist(data2, bins=30, histtype='step', label='Datensatz 2', color='orange', linewidth=1.5)

# Beschriftungen und Legende
plt.xlabel('AV')
plt.ylabel('Häufigkeit')
plt.title('Überlappende Histogramme')
plt.legend()

plt.show()
```
