```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from astropy.io import ascii, fits

b7 = pd.read_csv("B7.dat", sep=r"\s+")
b7.head()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(b7['N(C18O)'], bins=30, color='skyblue', edgecolor='black', alpha=0.7)

# Step 3: Customize the plot
plt.title('Histogram of C18O Density Distribution')
plt.xlabel('N(C18O)')
density=False
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
plt.figure(figsize=(9, 6))
sc = plt.scatter(b7["ra"], b7["dec"], c=b7["AV"], cmap='plasma', s=50, alpha=0.7)

plt.colorbar(sc, label="Extiction Value")
plt.xlabel("Rektaszension (RA)")
plt.ylabel("Deklination (Dec)")
plt.title("Extinction (C18O)")
plt.grid(False)
plt.show()
```
