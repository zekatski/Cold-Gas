### Molecular Hydrogen Generally and its Importance in the Universe

Molecular hydrogen ($H_2$) is the simplest molecule, composed of two hydrogen atoms, and was the first neutral molecule formed in the universe. Additionally, it is the most common molecule in the universe and plays a key role in the formation of stars and galaxies.

In the ISM (Interstellar Medium), molecular hydrogen is the primary component of cosmic molecular clouds. These clouds are large, dense regions and are among the coldest objects in the universe, with temperatures ranging from 10 to 50 Kelvin. In the coldest and densest corners of these clouds, star formation occurs. $H_2$ is a highly effective coolant at the low temperatures found in molecular clouds, which is essential as it allows the gas to compress further, eventually reaching densities that enable gravitational collapse and lead to star formation.

### Formation of Molecular Hydrogen
Although $H_2$ is the most abundant molecule in space, its formation is complex and not easily maintained. The issue with molecular hydrogen is that it is vulnerable to high-energy ultraviolet (UV) radiation, especially at wavelengths around 912-1100 angstroms (~90-110 nm). When $H_2$ molecules absorb UV photons, they are excited to higher energy states. If enough energy is absorbed, the bond between the two hydrogen atoms breaks, releasing them back into atomic form in a process known as photodissociation. This vulnerability creates the need for some form of protection against UV radiation to maintain stable levels of $H_2$. Additionally, the formation of molecular hydrogen itself is inefficient in the ISM without dust particles acting as catalysts.

##### Shielding of Molecular Hydrogen
A natural shield is provided by higher densities found in cosmic clouds, where a density greater than 100 particles/cm³ is necessary to counterbalance the rate of photodissociation. Dust grains—tiny solid particles composed of elements such as carbon, silicon, and metals—are concentrated in dense regions like cosmic clouds and act as an effective shield for molecular hydrogen. These particles are highly efficient at absorbing and scattering UV radiation, thus surrounding molecular hydrogen and absorbing much of the harmful UV light before it reaches the $H_2$ molecules.
But, $H_2$ can protect itself as well through a process called self-shielding, in which the outer layers of a molecular cloud absorb UV radiation and undergo dissociation, preserving the inner layers of $H_2$ from exposure.

##### $H_2$ Formation on Dust Particles
When a hydrogen atom collides with a dust grain and adheres to its surface, it can migrate across the surface to find another hydrogen atom. When two hydrogen atoms meet on the grain surface, they combine to form molecular hydrogen ($H_2$). This bonding process releases a small amount of energy, which is absorbed by the dust grain and dissipated as heat. The newly formed $H_2$ molecule then gains enough energy to escape from the dust particle.

##### Gas-Phase Reactions 
In very low-density regions, where dust grain catalysis is inefficient, alternative gas-phase reactions can form $H_2$, this process though is far less efficient and much rarer in the ISM. In this process, a proton (hydrogen ion)—usually created by cosmic rays—combines with a hydrogen atom to form an $H_2⁺$ ion. This $H_2⁺$ ion then reacts with either a free electron in the ISM or another hydrogen atom, resulting in the formation of neutral molecular hydrogen and a hydrogen ion.

Note: $H_2$ cannot generally form in the ISM through the direct collision of two hydrogen atoms. The reason behind this is that when two hydrogen atoms combine, energy is released as they transition from an excited state to a lower bound state. For the $H_2$ molecule to remain stable, this energy must be released or transferred; otherwise, the molecule will break apart again. In the ISM, there is virtually no means of dissipating this energy. When H₂ forms on a dust particle, the energy can be discharged into the grain, which makes dust essential for the formation of molecular hydrogen.

### Distribution of Molecular Hydrogen in the Milky Way
For understanding galactic evolution, molecular hydrogen is crucial. The amount and distribution of $H_2$ in a galaxy determine the galaxy's capacity to form new stars. Galaxies rich in $H_2$ are often actively forming stars, while galaxies with little $H_2$ tend to be older, less active, and often “quenched” in terms of star formation. Since the formation of $H_2$ occurs primarily in cosmic clouds, these clouds are naturally the main locations where molecular hydrogen can be found. Generally, molecular clouds are located in regions with increased density and mass. In a spiral galaxy like the Milky Way, this typically occurs in the spiral arms. The highest concentrations of $H_2$ are found in the inner regions of the galaxy, within about 5 kpc of the galactic center, where the disk is densest.

### Why is it so Hard to Detect $H_2$?
Due to many reasons, we would really like to know the exact distribution of molecular hydrogen clouds in our galaxy and others as well. Interestingly, in most cases, it is impossible for us to detect $H_2$ directly. The main issues are:

1. $H_2$ is a symmetrical molecule and lacks a permanent dipole moment, which prevents it from emitting or absorbing light in the radio and infrared wavelengths commonly used to observe other molecules.

2. Although $H_2$ can technically emit through quadrupole transitions (a higher-order effect), these transitions are extremely weak and occur only under specific conditions, such as high temperatures (over 100 K), where the molecule has enough energy to enter an excited state. But, these temperatures are not typically found in classical cosmic clouds.

3. $H_2$ can absorb UV light, and these absorptions can, in theory, be used to detect it. However, UV light is mostly absorbed by interstellar dust and gas, as explained earlier, making it difficult to observe $H_2$ directly in dense molecular clouds.

### The History of the Hunt for $H_2$ ###

#### Early Works and Ideas ####

In 1904, Johannes Hartmann observed the spectral lines of ionized calcium (Ca II) that appeared to remain unchanged despite the star’s motion. This finding led him to conclude that there was something between the light source and Earth that absorbed photons at a specific wavelength. The paper he published marks the first scientific evidence of the existence of clouds in the ISM. Later works of the early 20th century suggested that these interstellar clouds also consist of molecular components. In the following decades, further publications provided viable evidence for this idea. During the 1950s, the 21 cm line in radio spectra caused by atomic hydrogen was discovered, leading astronomers to search for a similar tracer of molecular hydrogen. The abundance of $H_2$ clouds was theorized in a paper published in 1951, but this very paper also discussed the potential difficulty of finding molecular hydrogen due to its physical properties.

#### Searching for Effective Tracers ####

Almost two decades later, in the late ’60s, new publications aimed to encourage scientists to shift their research objectives in a different direction. The development of an indirect $H_2$ tracing method was necessary. In the following years, a concept for the process of $H_2$ formation on dust particles was proposed, further supporting the idea that, in interstellar space, hydrogen is also abundant in its molecular form. During the ’70s, many new ideas and discoveries emerged, starting with a 1970 paper on observations of line radiation attributed to CO in cold interstellar gas, which introduced the molecule as a viable $H_2$ tracing candidate. This idea was successfully reinforced by papers published in the following years as CO was established as the primary tracer for molecular hydrogen. A notable theoretical advancement from this time is the CO-to-$H_2$ conversion factor, commonly denoted as XCO. The classic review paper on the evolution of CO as a tracer was published in 1986, providing an interesting view of the historical development of carbon monoxide as a tracer. Over the years, these carbon-monoxide-centered methods contributed to the search for interstellar molecular clouds and were continuously improved and refined until the 2000s. A more recent paper from 2013 gives a comprehensive review of CO as a tracer of $H_2$, also discussing the limitations and dependencies of this method. These limitations, mostly occurring in CO “dark” regions, have prompted a search for CO-unrelated tracing techniques.

#### Modern Methods and Alternatives to CO ####

With newer and more advanced instrumentation, the accuracy and range of modern surveys have drastically increased each year. This has resulted in a wider spectrum of available $H_2$ tracing methods, as well as an improved understanding of the role CO holds in locating molecular hydrogen. This more recent data has been gathered using a myriad of instruments and techniques. While radio and (sub)millimeter surveys manage to map molecular tracers other than CO, which are particularly helpful in regions of low CO abundance, infrared observations can locate dust emissions, which are also suggested indicators of $H_2$ presence. Additionally, studies in the gamma-ray range allow for the detection of dark gas clouds interacting with cosmic rays. A paper published in the 2015 Annual Review of Astronomy and Astrophysics provides a broad discussion and analysis of the Milky Way’s molecular clouds and addresses the various methods of tracing $H_2$ and the effectiveness of these tracers.

Using CO as a Proxy: 
Carbon monoxide (CO) is often used as a tracer for molecular hydrogen because it emits strongly in the radio wavelength (particularly at the J=1→0J=1→0 rotational transition of 2.6 mm) and is easier to detect. 
Since CO and H22​ often coexist in dense molecular clouds,astronomers use the ratio between CO intensity and H22​ column density, known as the CO-to-H22​ conversion factor (or XCOXCO​), to estimate H22​ densities. 
The value of XCOXCO​ can vary with cloud conditions, such as metallicity and radiation environment

UV Absorption Lines:
In regions near hot stars, molecular hydrogen can be detected through ultraviolet absorption lines in the Lyman and Werner bands, which occur around 912–1100 Å. 
Observations from instruments like the Hubble Space Telescope and the Far Ultraviolet Spectroscopic Explorer (FUSE) have been crucial in directly detecting H22​ via this method, especially in translucent and diffuse regions of the ISM

Infrared Emission Lines: 
Vibrational and rotational transitions of H22​ in the near-infrared, particularly at wavelengths of 2.12 µm, can be used to study warmer molecular hydrogen, such as in photodissociation regions (PDRs) near hot stars or shocked gas. 
Instruments like the Spitzer Space Telescope and ground-based observatories equipped with IR spectrometers have provided valuable data on H22​ in these environments

Gamma-Ray Observations: 
Cosmic rays interacting with H22​ in molecular clouds produce gamma-ray emissions that can be detected by telescopes like the Fermi Gamma-ray Space Telescope. 
This method, while indirect, allows estimation of molecular hydrogen in regions where CO is under-abundant, such as in low-metallicity environments where traditional methods may be less effective

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# plot
plt.figure(figsize=(12, 7))

# data filtering
pv_XC18O = (df['X(C18O)'] >= 0) & (df['X(C18O)'] <= 0.8e-6)  
data_C18O = df['X(C18O)'][pv_XC18O] 

# Histogramm
plt.hist(data_C18O, bins=100, density=True, color='skyblue', edgecolor='skyblue', alpha=0.8, label='Histogram')

# Gaussian distribution
mean_C18O = np.mean(data_C18O)  # mean
std_C18O = np.std(data_C18O)    # standard deviation
data_C18O_tn = data_C18O.count()
std_error_C18O = std_C18O/(data_C18O_tn**(0.5))
x_C18O = np.linspace(min(data_C18O), max(data_C18O), 1000)  # x-values for PDF
pdf_C18O = norm.pdf(x_C18O, mean_C18O, std_C18O)  # PDF of the normal distribution

# Plot of guassian distibution
plt.plot(x_C18O, pdf_C18O, color='red', lw=2, label='Gaussian Fit')

# mean and standard deviation lines
plt.axvline(mean_C18O, color='blue', linestyle='--', lw=2, label=f'Mean = {mean_C18O:.2e}')
plt.axvline(mean_C18O + std_C18O, color='darkgreen', linestyle='--', lw=2, label=f'+1σ = {mean_C18O + std_C18O:.2e}')
plt.axvline(mean_C18O - std_C18O, color='darkgreen', linestyle='--', lw=2, label=f'-1σ = {mean_C18O - std_C18O:.2e}')
plt.axvline(mean_C18O + 2.5*std_C18O, color='lime', linestyle='--', lw=2, label=f'+2.5σ = {mean_C18O+ 2.5*std_C18O:.2e}')

# axes and plot titel
plt.title('Distribution of X(C18O) with Gaussian Fit', fontsize=16)
plt.xlabel('X(C18O)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xlim(min(data_C18O), 0.8e-6)

# legend
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(fontsize=12)

# Layout 
plt.tight_layout()
plt.show()
```

```python
count_C18O = df['X(C18O)'].count()
lost_data_frc = data_C18O_tn/count_C18O
print(f"By cutting negative values and outliers we get to a fraction of {lost_data_frc:.2f}% of the original dataset.")
C18O_CI_l = mean_C18O - 2.5*std_error_C18O
C18O_CI_u = mean_C18O + 2.5*std_error_C18O
print(f"The confidence interval for the mean, for a 2.5σ range, lies between {C18O_CI_l:.2e} and {C18O_CI_u:.2e}")
```
