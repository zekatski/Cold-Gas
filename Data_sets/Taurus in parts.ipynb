```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from astropy.io import ascii, fits

b211 = pd.read_csv("B211.dat", sep=r"\s+")
b211.head(2000)
```

```python
b213 = pd.read_csv("B213.dat", sep=r"\s+")
b213.head()
```

```python
b7 = pd.read_csv("B7.dat", sep=r"\s+")
b7.head()
```

```python
b216 = pd.read_csv("B216.dat", sep=r"\s+")
b216.head()
```

```python
plt.figure(figsize=(9, 6))
sc = plt.scatter(b211["ra"], b211["dec"], c=b211["N(H2)"], cmap='plasma', s=50, alpha=0.7)

plt.colorbar(sc, label="Extiction Value")
plt.xlabel("RA")
plt.ylabel("Dec")
plt.title("Extinction of N(C18O) in B211")
plt.grid(False)
plt.show()
```

```python
plt.figure(figsize=(9, 6))
sc = plt.scatter(b216["ra"], b216["dec"], c=b216["AV"], cmap='plasma', s=50, alpha=0.7)

plt.colorbar(sc, label="Extiction Value")
plt.xlabel("RA")
plt.ylabel("Dec")
plt.title("Extinction of C18O in B216")
plt.grid(False)
plt.show()
```

```python
plt.figure(figsize=(9, 6))
sc = plt.scatter(b7["ra"], b7["dec"], c=b7["AV"], cmap='plasma', s=50, alpha=0.7)

plt.colorbar(sc, label="Extiction Value")
plt.xlabel("Rektaszension (RA)")
plt.ylabel("Deklination (Dec)")
plt.title("Extinction of C18O in B7")
plt.grid(False)
plt.show()
```

```python
plt.figure(figsize=(9, 6))
sc = plt.scatter(b213["ra"], b213["dec"], c=b213["N(C18O)"], cmap='plasma', s=50, alpha=0.7)

plt.colorbar(sc, label="N(C18O) Distribution")
plt.xlabel("RA")
plt.ylabel("Dec")
plt.title("B213 N(C18O) Distribution")
plt.grid(False)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(b211['N(C18O)'], bins=45, color='skyblue', edgecolor='black', alpha=0.7)

# Step 3: Customize the plot
plt.title('Histogram of C18O Density Distribution')
plt.xlabel('N(C18O)')
density=False
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
